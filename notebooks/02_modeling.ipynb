## Feature Engineering and Model Development

This notebook documents the feature engineering and model development stage for the Heart Disease UCI classification project.

The objective of this stage is to prepare machine learning features, build reproducible preprocessing pipelines, train multiple classification models, perform hyperparameter tuning, and compare model performance using appropriate classification metrics.

The assignment requires final machine learning features to include scaling and encoding, at least two trained classification models, model selection/tuning documentation, cross-validation, and evaluation using metrics such as accuracy, precision, recall, and ROC-AUC.

For this implementation, two models are evaluated:

1. Logistic Regression
2. Random Forest Classifier

Logistic Regression provides a simple and interpretable baseline model, while Random Forest provides a more flexible nonlinear model that can capture interactions between clinical features.

```python
import pandas as pd
from pathlib import Path

results_path = Path("../reports/model_comparison.csv")
results = pd.read_csv(results_path)

results
```

### Model Comparison

The models were compared using 5-fold stratified cross-validation on the training set and final evaluation on the held-out test set.

The primary model selection metric was ROC-AUC because it evaluates the model's ability to separate positive and negative classes across different classification thresholds. Accuracy, precision, recall, and F1-score were also reviewed to provide a more complete understanding of model performance.

Recall is especially important in this use case because false negatives may represent patients with heart disease risk who are incorrectly predicted as low risk. Precision is also relevant because excessive false positives could reduce confidence in model recommendations.

```python
best_model = results.sort_values(by="test_roc_auc", ascending=False).iloc[0]
best_model
```

### Selected Model

The final model was selected based on the highest test ROC-AUC score after hyperparameter tuning. The complete trained pipeline, including preprocessing and classifier, was saved as:

`model/heart_disease_pipeline.joblib`

Saving the full pipeline ensures that the same preprocessing steps are applied during training and inference. This improves reproducibility and reduces the risk of training-serving skew.
